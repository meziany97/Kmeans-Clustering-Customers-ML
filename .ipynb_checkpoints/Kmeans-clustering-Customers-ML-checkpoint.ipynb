# importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# importing the dataset 'customers-data'

```python
dataset = pd.read_csv("Customers-data.csv")
```

# defining the data of X for trainning the model

```python
# X = data.iloc[:,[3,4]].values
X = dataset.iloc[:,[3,4]].values
```

```python
#  import Kmeans Model from sklear library
from sklearn.cluster import KMeans
# Wcss : Within Cluster Sum of Squares
wcss=[]
# we will use the elbow methode to find the optimal number of clusers
for i in range (3,13):
    Kmeans = KMeans(n_clusters = i, init = "k-means++", max_iter = 300, n_init = 10, random_state =0)
    Kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
```

# plot the graph of elbow methode to find the optimal number vof clusters

```python
plt.plot(range(3,13),wcss)
plt.title('Elbow methode')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

# fiting the model on 5 clusters and make prediction

```python
KMeans = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 10, random_state =0)
y_means = KMeans.fit_predict(X)
```

# scattering our clusters on a plot

```python
plt.scatter(X[y_means == 0,0],X[y_means==0,1], s =100, c='red' , label='cluster 1')
plt.scatter(X[y_means == 1,0],X[y_means==1,1], s =100, c='blue' , label='cluster 2')
plt.scatter(X[y_means == 2,0],X[y_means==2,1], s =100, c='yellow' , label='cluster 3')
plt.scatter(X[y_means == 3,0],X[y_means==3,1], s =100, c='green' , label='cluster 4')
plt.scatter(X[y_means == 4,0],X[y_means==4,1], s =100, c='aqua' , label='cluster 5')
# plot the centroid of each clusters
plt.scatter(KMeans.cluster_centers_[:,0], KMeans.cluster_centers_[:,1], s = 200, c = 'orange', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income $')
plt.ylabel('spending Score (1-100)')
plt.legend()
plt.show()
```
